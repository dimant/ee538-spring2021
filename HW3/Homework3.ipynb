The small-signal model of the circuit can be discussed as 3 separate stages (transfer functions): 1) $H_1(s)$volage divider, 2) $H_2(s)$ common emitter, and 3) $H_3(s)$ emitter follower. The transfer function of the entire circuit is then $H(s)=H_1(s) \cdot H_2(s) \cdot H_3(s)$

Voltage divider:
\begin{equation}
H_{1}(s) = \frac{v_b}{v_{in}} = \frac{R_{b1} || R_{b2}}{R_{b1} || R_{b2} + \frac{1}{sC}}
\end{equation}

\begin{equation}
\frac{v_b}{v_{in}} = \frac{\frac{R_{b1}R_{b2}}{R_{b1} + R_{b2}}}{\frac{R_{b1}R_{b2}}{R_{b1} + R_{b2}} + \frac{1}{sC}} \\
\frac{v_b}{v_{in}} = \frac{sC_1R_{b1}R_{b2}}{R_{b2} + R_{b1}(1 + sC_1R_{b2})}
\end{equation}

\begin{equation}
\omega_0 = \frac{1}{(R_b1||R_b2)C}
\end{equation}

Common emitter:
\begin{equation}
r_{\pi} = \frac{\beta_0}{g_m}\\
Z_e = \frac{sCR_e}{R_e + sC} \\
v_{out} = -i_c R_c\\
i_e = i_b + i_c \\
v_{\pi} = v_b - v_e \\
i_c = -\frac{v_{out}}{R_c} = -g_m v_{\pi} = -g_m(v_b - v_e)
\end{equation}
express $v_e$
\begin{equation}
v_e = v_b - \frac{v_{out}}{g_mR_c}\\
i_b = \frac{v_b - v_e}{r_{\pi}} \\
\end{equation}
substitute $i_b,i_c$
\begin{equation}
i_e =  \frac{v_b - v_e}{r_{\pi}} - g_m(v_b - v_e)\\
i_e = (v_b - v_e) (\frac{g_m}{\beta_0} - g_m) \\
\end{equation}
substitute $v_e$
\begin{equation}
i_e = - \frac{v_{out}}{g_mR_c}\left(\frac{g_m}{\beta_0} - g_m\right) = - \frac{v_{out}}{R_c}\left(1 - \frac{1}{\beta_0}\right)\\
1 - \frac{1}{\beta_0} \approx 1 \\
v_e = - \frac{sCR_e}{R_e + sC}\frac{v_{out}}{R_c}
\end{equation}

\begin{equation}
-g_m (v_b - v_e) = - \frac{v_{out}}{R_c}\\
g_m v_b - g_m v_e = \frac{v_{out}}{R_c}\\
g_m v_b = \frac{v_{out}}{R_c} + g_m v_e \\
g_m v_b = \frac{v_{out}}{R_c} - g_m \frac{sCR_e}{R_e + sC}\frac{v_{out}}{R_c}\\
g_m v_b = v_{out}\left(\frac{1}{R_c} - g_m \frac{sCR_e}{R_e + sC}\frac{1}{R_c}\right) \\
\frac{v_{out}}{v_b}  = \frac{g_m}{\frac{1}{R_c} - g_m \frac{sCR_e}{R_e + sC}\frac{1}{R_c}} \\
H_2(s)=\frac{v_{out}}{v_b} = \frac{g_mR_c}{1 - \frac{sC_2g_mR_e}{sC_2+R_e}}\\
H_2(s) = -\frac{g_mR_c}{1 - g_mZ_e} \approx -\frac{R_c}{Z_e}
\end{equation}

Emitter follower:
\begin{equation}
v_{out}=v_e \\
i_c = \beta_0 i_b\\
g_m (v_{in} - v_e) = \beta_0 i_b \\
i_e = i_b + i_c = i_b (\beta_0 + 1) \\
v_{out} = i_b (\beta_0 + 1) R_e \\
v_{in} = i_b \frac{\beta_0}{g_m} \\
\frac{v_{out}}{v_{in}} = \frac{i_b (\beta_0 + 1) R_e}{i_b \frac{\beta_0}{g_m}}\\
\frac{v_{out}}{v_{in}} = g_m\frac{(\beta_0 + 1) R_e}{\beta_0} \\
\frac{v_{out}}{v_{in}} = g_m(1 + \frac{1}{\beta_0}) R_e \\
1 + \frac{1}{\beta_0} \approx 1 \\
\frac{v_{out}}{v_{in}} = g_mR_e
\end{equation}

Transfer function:

\begin{equation}
H(s) = -\frac{sC_1R_{b1}R_{b2}}{R_{b2} + R_{b1}(1 + sC_1R_{b2})}\frac{g_mR_c}{1 - g_mZ_e}g_mR_{e2}
\end{equation}

\begin{equation}
g_m = \frac{I_C}{V_T}\\
R_{out} = R_E || \frac{1}{g_m} = R_E || \frac{V_T}{I_C} = \frac{R_E \frac{V_T}{I_C}}{R_E + \frac{V_T}{I_C}}\\
R_E = -\frac{R_{out} V_T}{I_C R_{out}-V_T}
\end{equation}


$R_s$ for the emitter follower is output resistance of common emitter stage. Output resistance of common emitter is $R_C$.


```python
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import ticker, cm
import numpy as np

def plot_bjt_NF(Imin, Imax):
    fig, ax = plt.subplots(figsize=(10.0, 7.5))

    k = 1.38e-23
    T = 300
    q = 1.602e-19
    V_T = k*T/q

    beta = 250
    R_s = 100 # Ohm
    r_b = 2 # Ohm
    R_out = 50 #Ohm
    Vcc = 9 # V
    G = 100 # 40dB
    G2 = G**2
    
    ic1 = np.logspace(np.log10(Imin), np.log10(Imax), num = 1000)
    ic2 = np.logspace(np.log10(Imin), np.log10(Imax), num = 1000)
    I_C1, I_C2 = np.meshgrid(ic1, ic2)

    R_s2 = R_s**2
    R_C = Vcc / I_C1
    R_E = R_C / G
    g_m = I_C1 / V_T
    g_m2 = g_m**2
    
    R_sef2 = R_C**2
    g_mef = I_C2 / V_T
    g_mef2 = g_mef**2
    R_E2 = R_out * V_T / (V_T - I_C2 * R_out)
    
    i_nR_E2 = 4 * k * T / R_E2
    i_nI_C22 = 2 * q * I_C2
    i_nbef2 = 2 * q * I_C2 / beta
    e_nbef2 = 4 * k * T * r_b
    
    e_nef2 = i_nbef2*R_sef2 + e_nbef2 + (i_nR_E2 + i_nI_C22)/g_mef2
    
    i_nR_C2 = 4 * k * T / R_C
    i_nR_E2 = 4 * k * T / R_E
    i_nI_C12 = 2 * q * I_C1
    i_nbce2 = 2 * q * I_C1 / beta
    e_nbce2 = 4 * k * T * r_b
    
    # voltage divider is not included because its noise contribution is not current dependent
    e_nce2 = i_nbce2 * R_s2 + e_nbce2 + (e_nef2/G2) + (i_nR_C2 + i_nR_E2 + i_nI_C12)/g_m2
        
    NF = 1 + e_nce2/(4*k*T*R_s)
    
    cp = ax.contourf(I_C1, I_C2, 10*np.log10(NF), levels=np.linspace(0,100, num=25))
    plt.xlabel(r'Collector Current $I_{C1}$ [A]')
    plt.ylabel(r'Collector Current $I_{C2}$ [A]')
    plt.xscale('log')
    plt.yscale('log')
    fig.colorbar(cp)
    
    min_NF = np.amin(NF)
    idx = np.where(NF == min_NF)
    
    display(f'Minimum noise figure: {min_NF}, {10*np.log10(min_NF)} dB')
    display(f'I_C1 {ic1[idx[0]]} I_C2 {ic2[idx[1]]}')

# note that 2SC3324 is maximum rated to 0.1A
plot_bjt_NF(1e-11, 1)
plot_bjt_NF(1e-6, 1)
```

$I_{C1}$ influences the noise figure stronger than $I_{C2}$ because the noise of the emitter-follower stage is input referred to the input of the circuit via the voltage gain of the amplifying stage. Past $10^{-1}$Amp the noise figure starts rising again because noise due to $r_b$ starts dominating other sources.

Amplifier filter circuit:


```python
def circuit(I_C1, I_C2):
    k = 1.38e-23
    T = 300
    q = 1.602e-19
    V_T = k*T/q
    
    beta = 250
    R_s = 100 # Ohm
    r_b = 2 # Ohm
    R_out = 50 #Ohm
    Vcc = 9 # V
    G = 100 # 40dB
    f_c = 70 # Hz
    hr = 0.65 # V headroom
    
    V_b = hr + (Vcc - hr) / 2
    R_B1 = 1000 # Ohm
    R_B2 = V_b * R_B1 / (Vcc - V_b)
    R_B = R_B1 * R_B2 / (R_B1 + R_B2)
    C_AC1 = 1 / (1 * np.pi * f_c * R_B)
    
    R_C = Vcc / I_C1
    R_E = R_C / G
    C_AC2 = 1 / (2 * np.pi * f_c * R_E)
 
    R_E2 = R_out * V_T / (V_T - I_C2 * R_out)
    
    display(f'V_b {V_b} V, R_B1 {R_B1} Ohm, R_B2 {R_B2} Ohm, R_B {R_B} Ohm, C_AC1 {C_AC1} F, R_C {R_C} Ohm, R_E {R_E} Ohm, C_AC2 {C_AC2} F, R_E2 {R_E2} Ohm')

circuit(0.0004453, 10e-5)
```
